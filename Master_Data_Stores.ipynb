```python
def latlon(point): # Function to get latitude and longitude from store location

    c = point.split("(")
    c = c[1].split(" ")
    lon = c[0]
    lat = c[1].split(")")[0]
    return lat,lon

from urllib.request import urlopen 
import json

def getplace(lat, lon): # function to get street,city,county and zipcode from latitude and longitude
    key = "AIzaSyD7yd0T_JLTzzLAbUEoIpREBXFMRtlq26M"
    url = "https://maps.googleapis.com/maps/api/geocode/json?"
    url += "latlng=%s,%s&sensor=false&key=%s" % (lat, lon, key)
    v = urlopen(url).read()
    j = json.loads(v)
    components = j['results'][0]['address_components']
    try:
        for c in components:
            if "locality" in c['types']:
                city = c['long_name']
            if "administrative_area_level_2" in c['types']:
                 county= c['long_name']
                 county = county.split(' ')
                 county = county[0]
            if "establishment" in c['types']:
                 street = c['long_name']
            if "postal_code" in c['types']:
                 zipcode = c['long_name']
        place = (street,city,county,zipcode)             
    except:
        place = None
    return place

```
